```python
#numpy and pandas used for data manipulation
import numpy as np
import pandas as pd
#matplotlib used for visualization
from matplotlib import pyplot as plt
#os used to get file directories
import os

from sklearn.decomposition import PCA
import cv2
import random
```

```python
allnames = []
straight2 = []
straight1 = []
for (dirpath, dirnames, filenames) in os.walk("faces/"):
    allnames.extend(filenames)
for name in allnames[1:]:
    if name.split("_")[1] == "straight":
        if name[-5:] == '2.pgm':
            straight2.append('faces/' + name.split('_')[0] + '/' + name)
        elif name[-5:] == '4.pgm':
            pass
        elif name[-3:] =='pgm':
            straight1.append('faces/' + name.split('_')[0] + '/' + name)
```

```python
img1 = cv2.imread(straight1[0],0)
plt.imshow(img1, plt.cm.gray)
plt.show()
```

```python
img2 = cv2.imread(straight2[0],0)
plt.imshow(img2, plt.cm.gray)
plt.show()
```
