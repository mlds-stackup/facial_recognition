```python
#numpy and pandas used for data manipulation
import numpy as np
import pandas as pd
#matplotlib used for visualization
from matplotlib import pyplot as plt
#os used to get file directories
import os
#opencv used to make get images into numbered arrays
import cv2
import random
from time import time

from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_lfw_people
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
```

First, grab all the file directories for images which are size 1

```python
def get_filedir(left, right, straight, up):
    allnames = []
    count = 0
    tempname = 'an2i'
    for (dirpath, dirnames, filenames) in os.walk("faces/"):
        allnames.extend(filenames)
    for name in allnames[1:]:
        if (name.split('_')[-1] not in ['2.pgm', '4.pgm', 'sunglasses.pgm']) and (name[-3:] =='pgm'):
            if name.split('_')[0] != tempname:
                count = count + 1
                tempname = name.split('_')[0]
                left.append([])
                right.append([])
                straight.append([])
                up.append([])
            if name.split('_')[1] == "left":
                left[count].append('faces/' + name.split('_')[0] + '/' + name)
            elif name.split('_')[1] == "right":
                right[count].append('faces/' + name.split('_')[0] + '/' + name)
            elif name.split('_')[1] == "straight":
                straight[count].append('faces/' + name.split('_')[0] + '/' + name)
            elif name.split('_')[1] == "up":
                up[count].append('faces/' + name.split('_')[0] + '/' + name)
```

split the set into a training set and a testing set, with pics organized per person, and 30% in test set

```python
def get_person_name(name_list, filedir):
    for person in filedir:
        for img in person:
            name_list.append(img.split('/')[1])
```

```python
name_s = []
get_person_name(name_s, straight)
```

```python
def get_img_matrix(img_list, filedir):
    for person in filedir:
        for img in person:
            img_list.append(cv2.imread(img, 0).ravel())
```

```python
s_img_list = []
get_img_matrix(s_img_list, straight)
```

```python
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(s_img_list, name_s, test_size=0.3, random_state=0)
```

## now applying the above to all profiles

```python
name_r = []
name_l = []
name_u = []

r_img_list = []
l_img_list = []
u_img_list = []


get_person_name(name_r, right)
get_img_matrix(r_img_list, right)

get_person_name(name_l, left)
get_img_matrix(l_img_list, left)

get_person_name(name_u, up)
get_img_matrix(u_img_list, up)


X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(r_img_list, name_r, test_size=0.3, random_state=0)
X_train_l, X_test_l, y_train_l, y_test_l = train_test_split(l_img_list, name_l, test_size=0.3, random_state=0)
X_train_u, X_test_u, y_train_u, y_test_u = train_test_split(u_img_list, name_u, test_size=0.3, random_state=0)

pca_r = PCA(n_components=50)
pca_r.fit(X_train_r)

pca_l = PCA(n_components=50)
pca_l.fit(X_train_l)

pca_u = PCA(n_components=50)
pca_u.fit(X_train_u)
```

```python
x_train_pca = np.concatenate((pca_r.transform(X_train_r), pca_l.transform(X_train_l), pca_s.transform(X_train_s), pca_u.transform(X_train_u)), axis=0)
x_test_pca = np.concatenate((pca_r.transform(X_test_r), pca_l.transform(X_test_l), pca_s.transform(X_test_s), pca_u.transform(X_test_u)), axis=0)

y_train = y_train_r + y_train_l + y_train_s + y_train_u
y_test = y_test_r + y_test_l + y_test_s + y_test_u
```

```python
print("Fitting the classifier to the training set")
t0 = time()
param_grid = {'C': [1e3, 5e3, 1e4, 5e4, 1e5],
              'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.1], }
clf = GridSearchCV(SVC(kernel='rbf', class_weight='balanced'), param_grid)
clf = clf.fit(x_train_pca, y_train)
print("done in %0.3fs" % (time() - t0))
print("Best estimator found by grid search:")
print(clf.best_estimator_)
```

```python
print("Predicting people's names on the test set")
t0 = time()
y_pred = clf.predict(x_test_pca)
print("done in %0.3fs" % (time() - t0))

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))#, labels=range(n_classes)))
```
