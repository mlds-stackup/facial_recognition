```python
#numpy and pandas used for data manipulation
import numpy as np
import pandas as pd
#matplotlib used for visualization
from matplotlib import pyplot as plt
#os used to get file directories
import os

from sklearn.decomposition import PCA
import cv2
import random
```

```python
f = []
for (dirpath, dirnames, filenames) in os.walk("faces/"):
    f.extend(filenames)
```

```python
faces4 = []
faces2 = []
faces1 = []
for name in f:
    if name[-5:] == '2.pgm':
        faces2.append('faces/' + name.split('_')[0] + '/' + name)
    elif name[-5:]== '4.pgm':
        faces4.append('faces/' + name.split('_')[0] + '/' + name)
    elif name[-3:] =='pgm':
        faces1.append('faces/' + name.split('_')[0] + '/' + name)
```

```python
img1 = cv2.imread(faces1[0],0)
pyplot.imshow(img1, pyplot.cm.gray)
pyplot.show()
```

```python
img2 = cv2.imread(faces2[0],0)
pyplot.imshow(img2, pyplot.cm.gray)
pyplot.show()
```

```python
img4 = cv2.imread(faces4[0],0)
pyplot.imshow(img4, pyplot.cm.gray)
pyplot.show()
```
